```python
import torch
import torch.nn as nn

class TinyMLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(3, 4) # nn.Linear(3, 4) 只改变最后一个维度的特征数量（把 3 变成 4），而不改变你输入数据的条数（N保持不变）。
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(4, 1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

model = TinyMLP()

for name, param in model.named_parameters():
    print(name, param.shape)
```

输入 x
→ 模型预测 pred = model(x)
→ 计算损失 loss
→ 清空旧梯度 zero_grad()
→ 反向传播 loss.backward()
→ 更新参数 optimizer.step()

```python
import torch
import torch.nn as nn
import torch.optim as optim

# 1. 准备数据
x = torch.tensor([[1.0], [2.0], [3.0], [4.0]])
y = torch.tensor([[2.0], [4.0], [6.0], [8.0]])

# 2. 定义模型
model = nn.Linear(1, 1)

# 3. 定义损失函数
criterion = nn.MSELoss()

# 4. 定义优化器
optimizer = optim.SGD(model.parameters(), lr=0.1)

# 5. 训练
for epoch in range(100):
    pred = model(x)                  # 前向传播
    loss = criterion(pred, y)        # 计算损失

    optimizer.zero_grad()            # 清空旧梯度
    loss.backward()                  # 反向传播
    optimizer.step()                 # 更新参数

    if epoch % 10 == 0:
        print(epoch, loss.item())
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

x = torch.tensor([[1.0], [2.0], [3.0]])
y = torch.tensor([[2.0], [4.0], [6.0]])

model = nn.Linear(1, 1)
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

pred = model(x)
loss = criterion(pred, y)

print("训练前参数：")
for name, param in model.named_parameters():
    print(name, param.data)

optimizer.zero_grad()
loss.backward()

print("\n梯度：")
for name, param in model.named_parameters():
    print(name, param.grad)

optimizer.step()

print("\n更新后参数：")
for name, param in model.named_parameters():
    print(name, param.data)
```
